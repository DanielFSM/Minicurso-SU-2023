## Potencial de Lennard-Jones

O potencial de Lennard-Jones é dado por

$w(r)=-\frac{A}{r^6}+\frac{B}{r^{12}}$

Vamos lançar mão de um gráfico do potencial entre duas partículas que obedecem esta expressão. Alguns valores típicos dos parâmetros $A$ e $B$ para moléculas pequenas é da ordem de $A = 6.022\times10^{-3}kJ\cdot mol^{-1}\cdot nm^6$ e $B = 6.022\times10^{-6}kJ\cdot mol^{-1}\cdot nm^{12}$. Vamos construir um gráfico utilizando estes valores para distâncias $0.1nm<r<2nm$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A1 = 6.022*10**-3 #em kJ/mol m**6
B1 = 6.022*10**-6 #em J/mol m**12

A2 = 2*6.022*10**-3 #em kJ/mol m**6
B12 = 3*6.022*10**-6 #em J/mol m**12



def w1(x):
  return -(A1/(x**6))+(B1/(x**12))

def w2(x):
  return -(A2/(x**6))+(B2/(x**12))



# Create a NumPy array of x-values
x = np.linspace(0.2,2, 200)

# Create a figure
fig = plt.figure()

# Plot the curve for each value of n
y1 = w1(x)
y2 = w2(x)
plt.plot(x, y1, 'bo', linestyle='-',label=str('Potencial de LJ'))
plt.plot(x, y1, 'go', linestyle='-',label=str('Potencial de LJ'))

plt.xlabel('r (nm)')
plt.ylabel('w(r) (kJ/mol)')
plt.ylim(-3,2.0)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

A força entre as partículas seria,

$F(r)=-\frac{dw(r)}{dr}=-\frac{6A}{r^7}+\frac{12B}{r^{13}}$

Podemos avaliar graficamente como esta força varia com a separação $r$ entre as partículas.

```python
def f(x):
  return -((6*A)/(x**7))+((12*B)/(x**13))

# Create a NumPy array of x-values
x = np.linspace(0.2,2, 200)

# Create a figure
fig = plt.figure()

# Plot the curve for each value of n
y2 = f(x)
plt.plot(x, y2, 'ro',linestyle='-',label=str('Força Potencial de LJ'))

plt.xlabel('r (nm)')
plt.ylabel('F(r) ($10^{-11}$ N)')
plt.ylim(-15,2.0)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Podemos sobrepor as duas curvas

```python
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(x, y1, 'r-', label='Potencial de LJ')
ax2.plot(x, y2, 'b-', label='Força Potencial de LJ')
ax1.axhline(0.0, linestyle='--')
ax1.axvline(0.36, linestyle='--')

ax1.set_ylim([-2, 2])
ax2.set_ylim([-15, 15])


ax1.set_xlabel('r (nm)')
ax1.set_ylabel('w(r) (kJ/mol)')
ax2.set_ylabel('F(r) ($10^{-11}$ N)')

ax1.legend(loc='upper right')
ax2.legend(loc='lower right')

plt.show()
```
